```python
import sys
import os, glob
import csv 
import calendar
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
from datetime import datetime, timedelta
import pandas as pd
import xarray as xr
import netCDF4 as nc
```

```python
### Readin satellite proxy data
in_file = './data/Satellite_Proxies_WREF.csv'
df_sat_wref = pd.read_csv(in_file,sep=',',index_col=0,parse_dates=True,header=0)

in_file = './data/Satellite_Proxies_ABBY.csv'
df_sat_abby = pd.read_csv(in_file,sep=',',index_col=0,parse_dates=True,header=0)
```

```python
plt.figure(num=None, figsize=(12, 9), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 16})

st = pd.datetime(2021,3,1) ; en = pd.datetime(2021,10,1)
heatdome = pd.date_range(datetime(2021,6,25),datetime(2021,6,29),freq='D')

plt.subplot(3,1,1)
plt.plot(df_sat_wref.kNDVI[st:en],color='sienna',label = 'WREF')
plt.plot(df_sat_abby.kNDVI[st:en],color='cadetblue',label = 'ABBY')
plt.fill_between(heatdome, np.tile(0,len(heatdome)),np.tile(1,len(heatdome)),facecolor='peru',alpha=0.4)
plt.ylabel('kNDVI')
plt.legend()

plt.subplot(3,1,2)
plt.plot(df_sat_wref.NIRv[st:en],color='sienna')
plt.plot(df_sat_abby.NIRv[st:en],color='cadetblue')
plt.fill_between(heatdome, np.tile(0,len(heatdome)),np.tile(0.5,len(heatdome)),facecolor='peru',alpha=0.4)
plt.ylabel('NIRv')

plt.subplot(3,1,3)
plt.plot(df_sat_wref.Tropomi_SIF[st:en],color='sienna')
plt.plot(df_sat_abby.Tropomi_SIF[st:en],color='cadetblue')
plt.fill_between(heatdome, np.tile(0,len(heatdome)),np.tile(2,len(heatdome)),facecolor='peru',alpha=0.4)
plt.ylabel('Tropomi_SIF')

plt.tight_layout()
plt.savefig('./figs/Satellite_proxies_2021_WREF-ABBY.png',dpi=300)
```
